### Spectroscopy galaxy surveys:

Here we average over all realization of the spectroscopy survey. the stat is used in the cell above. 

```python
from lila import stats
importlib.reload(stats)
def average_gal_stat(savefile, num_maps=30):
    st = stats.Stats(kmin=1/250, z=2.499999947500001, vol_ratio=3)
    st_av = stats.Stats(kmin=1/250, z=2.499999947500001, vol_ratio=1)
    st.load_stat(f'/home/mahdi/HD2/LIM/gal_maps_spec/astrid_Rz7e-4_stats_masscut11.9_gaussianized_peak_kcut0.03_n0.hdf5')
    st_av.gal_pk, st_av.lim_pk, st_av.lim_gal_pk  = {}, {}, {}
    for key in ['k', 'power','modes']:
        st_av.lim_pk[key] = st.lim_pk[key]
        st_av.gal_pk[key] =  st.gal_pk[key]
        if key!='modes':
            st_av.lim_gal_pk[key] = st.lim_gal_pk[key]
    st_av.sigma_gal_pk = st.sigma_gal_pk
    st_av.sigma_lim_pk = st.sigma_lim_pk
    st_av.sigma_lim_gal_pk = st.sigma_lim_gal_pk
    st_av.lim_gal_sn = st.lim_gal_sn

    for i in range(1,num_maps):
        st = stats.Stats(kmin=1/250, z=2.499999947500001, vol_ratio=1)
        st.load_stat(f'/home/mahdi/HD2/LIM/gal_maps_spec/astrid_Rz7e-4_stats_masscut11.9_gaussianized_peak_kcut0.03_n{i}.hdf5')
        st_av.gal_pk['power'] += st.gal_pk['power']
        st_av.lim_pk['power'] += st.lim_pk['power']
        st_av.lim_gal_pk['power'] += st.lim_gal_pk['power']
        st_av.sigma_gal_pk += st.sigma_gal_pk
        st_av.sigma_lim_pk += st.sigma_lim_pk
        st_av.sigma_lim_gal_pk += st.sigma_lim_gal_pk
        st_av.lim_gal_sn += st.lim_gal_sn
    print(st.n3D)
    st_av.n3D = st.n3D
    st_av.gal_pk['power'] /= num_maps
    st_av.lim_pk['power'] /= num_maps
    st_av.lim_gal_pk['power'] /= num_maps
    st_av.sigma_gal_pk /= num_maps
    st_av.sigma_lim_pk /= num_maps
    st_av.sigma_lim_gal_pk /= num_maps
    st_av.lim_gal_sn /= num_maps

    st_av.save_stat(savefile=savefile)

average_gal_stat(savefile='/home/mahdi/HD2/LIM/gal_maps_spec/astrid_Rz7e-4_stats_masscut11.9_gaussianized_peak_kcut0.03_average.hdf5')
```

```python
from os import stat
from lila import stats
importlib.reload(stats)

def avergae_pkperp(pkxy, kx, ky):
    """P(k_x,k_y) to P(k_perp)"""
    print(pkxy.shape)
    pkperp = np.array([])
    kx2, ky2 = np.meshgrid(kx**2, ky**2, indexing='ij')
    kperp2 = kx2+ky2
    kperp2_bins = np.linspace(np.min(kperp2), np.max(kperp2), 10)
    bins = np.digitize(kperp2, kperp2_bins)
    for i in range(0, kperp2_bins.size):
        ind = np.where(bins == i)
        pkperp = np.append(pkperp, np.mean(pkxy[ind]))
    kperp_bins = np.sqrt(kperp2_bins)
    print(pkperp, kperp_bins)
    return pkperp, kperp_bins

def get_Cq(pk_file, boxsize):
    with h5py.File(pk_file,'r') as f:
        # Load the 3D power and clear the  zero mode, k=0
        kx, ky = f['x'][1::]*2*np.pi/boxsize[0], f['y'][1::]*2*np.pi/boxsize[1]
        pk3d = f['pk3d'][1::,1::,1::]
        print(kx)
        print(f['z'][:])
        # Avergae the 3D power along z-direction
        # Also clear the zero mode
        Cq = np.average(pk3d, axis=2)/boxsize[2]
        Cq, k= avergae_pkperp(Cq, kx, ky)
    return Cq, k

def plot_Cq(pk_files=['/run/media/mahdi/HD2/LIM/LATIS_source_power.hdf5']):
    labels = ['Full map', 'Upper half', 'lower half']
    Lzs =[485, 242, 242]
    colors = ['C0','C1','C2','C3']
    dperp = [2.5,3.7,10,13]
    fig, ax = plt.subplots(1,1, figsize=(10,5))
    i = 0
    for pk_file, label, Lz in zip(pk_files, labels, Lzs):
        boxsize = [94,54,Lz]
        Cq, k= get_Cq(pk_file, boxsize)
        ax.plot(k, Cq*(1/dperp[i]**2), label=label)
        ax.plot(k, [1]*k.size, ls='--')
    ax.set_ylabel(r'$C_q \ \bar{n}_{2D}$')
    ax.set_xlabel(r'$k_{\perp} [ h/cMpc] $')
    ax.set_yscale('log')
    ax.set_xscale('log')
    ax.grid(which='both', axis='both')
    ax.legend(framealpha=0)
    ax.set_xlim((0.01,0.1))
    ax.set_xticks(np.arange(0.01,0.101,0.01))
    ax.set_xticklabels(['0.01','','','','','','','','','0.1'])

    #ax.set_ylim(1e-8,1e-4)
    fig.tight_layout()

plot_Cq()
```

Order of magnitude calcualtions:

```python
fig, ax = plt.subplots(1,1)
blya, bqso = 0.2, 3.64
n_qso = 1e-6
labels = ['eBOSS','DESI','PFS-faint']

for l,st in zip(labels, [st_eboss_gaus, st_desi_gaus, st_latis_gaus]):
    p_lya_noise =  st.lya_noise_pk
    snr_ratio = (blya/bqso)*np.sqrt((1/n_qso)/(p_lya_noise))
    ax.plot(st.lya_pk['k'], snr_ratio, label=l)
ax.set_xscale('log')
ax.grid(which='both', axis='both')
ax.set_ylabel(r'$SNR_{Lya} / SNR_{Lya, DESI}$')
ax.set_xlabel('k[h/cMpc]')
ax.legend(framealpha=0)
#ax.set_ylim((0,25))
```
